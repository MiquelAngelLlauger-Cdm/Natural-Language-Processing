## 1. Jaccard Similarity from Scratch

This function tokenizes two strings, converts them to lowercase, and calculates the intersection over the union of their word sets.
This is used as a similarity feature between question pairs.

```python
q1 = "How do I learn NLP?"
q2 = "What is the best way to learn NLP?"
print(f"Jaccard Similarity: {jaccard_similarity_scratch(q1, q2):.4f}")
```

## 2. Advanced Feature Vector

The `get_advanced_features` function combines the basic CountVectorizer features with:
1. Jaccard Similarity (scratch implementation).
2. Relative length difference between questions.

This provides more context to the Logistic Regression model, allowing it to distinguish between questions with similar words but different structures.

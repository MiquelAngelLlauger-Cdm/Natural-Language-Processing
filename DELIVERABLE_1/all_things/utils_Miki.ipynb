# Miki's Utilities Explanation

In this notebook, I'll explain my advanced TF-IDF approach.

## Idea
Standard Bag-of-Words just counts word overlaps. By using TF-IDF (Term Frequency - Inverse Document Frequency) on character n-grams, we can capture sub-word structures and assign higher weights to rare, distinguishing sub-strings. We then compute the Cosine Similarity between the two TF-IDF vectors for each question pair.

```python
import pandas as pd
from utils import extract_miki_features

df_sample = pd.DataFrame({
    'question1': ["How do I learn Python?", "What is the capital of France?"],
    'question2': ["What is the best way to learn Python?", "Where is France?"]
})

from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

q1_sample = df_sample['question1'].fillna('').astype(str).values
q2_sample = df_sample['question2'].fillna('').astype(str).values
sample_qs = np.concatenate((q1_sample, q2_sample))

tfidf_word = TfidfVectorizer(analyzer='word', stop_words='english')
tfidf_word.fit(sample_qs)

tfidf_char = TfidfVectorizer(analyzer='char', ngram_range=(1, 3))
tfidf_char.fit(sample_qs)

features = extract_miki_features(df_sample, tfidf_word, tfidf_char)
print("Miki's TF-IDF Cosine Similarity Features:")
print(features)
```

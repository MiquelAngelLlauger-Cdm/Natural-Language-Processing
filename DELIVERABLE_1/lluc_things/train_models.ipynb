# Train models
This notebook trains a simple baseline (TF-IDF + Logistic Regression) and stores the model and vectorizer in the `models/` folder.
It uses functions from `utils.py` so keep this notebook clean of heavy function definitions.

```python
import os
from utils import load_quora, split_quora, prepare_pair_texts, build_tfidf, transform_tfidf, train_logreg, save_model, set_seed
import numpy as np
set_seed(123)
HOME = os.path.expanduser('~')
DATA_PATH = os.path.join(HOME, 'Datasets', 'QuoraQuestionPairs', 'quora_data.csv')
print('Data path:', DATA_PATH)
```

```python
# Load data
df = load_quora(DATA_PATH)
print('Loaded rows:', len(df))
```

```python
# Create splits as requested in the assignment
train_df, val_df, test_df = split_quora(df, seed=123)
print('train_df.shape=', train_df.shape)
print('val_df.shape=', val_df.shape)
print('test_df.shape=', test_df.shape)
```

```python
# Prepare texts and build TF-IDF on training set
train_texts = prepare_pair_texts(train_df)
val_texts = prepare_pair_texts(val_df)
test_texts = prepare_pair_texts(test_df)
vec, X_train = build_tfidf(train_texts)
X_val = transform_tfidf(vec, val_texts)
X_test = transform_tfidf(vec, test_texts)
```

```python
# Train a logistic regression baseline and save the model + vectorizer
y_train = train_df['is_duplicate']
MODEL_PATH = os.path.join('models', 'logreg_baseline.pkl')
VEC_PATH = os.path.join('models', 'tfidf_vectorizer.pkl')
if not os.path.exists(MODEL_PATH):
    clf = train_logreg(X_train, y_train, seed=123)
    save_model(clf, MODEL_PATH)
    save_model(vec, VEC_PATH)
    print('Saved model and vectorizer to models/')
else:
    print('Model already exists at', MODEL_PATH)
```

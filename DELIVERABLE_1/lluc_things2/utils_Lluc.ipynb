In this notebook, I'll explain the feature extraction functions I created in `utils.py` for Deliverable 1.

## Idea
My hypothesis is that duplicate questions often start with the same question word (like "what", "how", "why", etc.).
If two questions have high word overlap but one starts with "who" and the other with "where", they are likely asking different things and are not duplicates.

I added the following features to capture this:
1. `same_first_word`: Binary feature indicating if both questions start with the same word.
2. `q1_is_wh` and `q2_is_wh`: Binary features indicating if the first word of each question is a common WH-question word ('who', 'what', 'where', 'when', 'why', 'how', 'which', 'whom', 'whose').

```python
from utils import get_first_word, extract_lluc_features
import pandas as pd
```

### `get_first_word`
This function cleans the text and extracts the first word of the sentence.

```python
q1 = "What is the capital of France?"
print("First word of q1:", get_first_word(q1))

q2 = "how do I learn Python?"
print("First word of q2:", get_first_word(q2))
```

### `extract_lluc_features`
This function combines the baseline features (overlap, Jaccard, length diff) with my new first-word features.

```python
df_sample = pd.DataFrame({
    'question1': ["What is the meaning of life?", "Who is the president?"],
    'question2': ["What does life mean?", "Where is the president?"]
})
features = extract_lluc_features(df_sample)
print("Extracted Features Array:")
print(features)
```

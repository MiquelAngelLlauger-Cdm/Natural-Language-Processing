## Idea
Standard word overlap metrics are very rigid. If a word is misspelled or slightly altered, it counts as a mismatch. Drawing inspiration from the string distance theory materials, I used the **Levenshtein Edit Distance**. This metric calculates the minimum number of single-character edits required to change one question into the other. I then normalized it by the maximum length of the two strings to provide a consistent similarity feature.

```python
import pandas as pd
from utils import extract_jose_features

df_sample = pd.DataFrame({
    'question1': ["What is the capital of France?", "kitten"],
    'question2': ["What is the capitl of France?", "sitting"]
})

features = extract_jose_features(df_sample)
print("Jose's Normalized Edit Distance Features:")
print(features)
```

## Idea
Standard Bag-of-Words just counts word overlaps. By using TF-IDF (Term Frequency - Inverse Document Frequency) on character n-grams, we can capture sub-word structures and assign higher weights to rare, distinguishing sub-strings. We then compute the Cosine Similarity between the two TF-IDF vectors for each question pair.

```python
import pandas as pd
from utils import extract_miki_features

df_sample = pd.DataFrame({
    'question1': ["How do I learn Python?", "What is the capital of France?"],
    'question2': ["What is the best way to learn Python?", "Where is France?"]
})

features = extract_miki_features(df_sample)
print("Miki's TF-IDF Cosine Similarity Features:")
print(features)
```
